### 1. Working with individual stocks

This section demonstrates the usage of XQuantiPy module when working with individual stocks.

```python
# Import the required module
# !pip install nbformat
from xquantipy.stocks.ticker import Ticker
```

```python
# Getting individual stock object
AAPL = Ticker('AAPL', '1Y')
AAPL.data
```

```python
# Get the adjusted close of the stock
AAPL.get_adj_close()
```

```python
# Get the beta value of a stock
AAPL.get_beta(index="^GSPC")
```

```python
# Get the alpha value of a stock
AAPL.get_alpha(index="^GSPC", risk_free_rate=0.05)
```

```python
# Plot the adjust close chart
AAPL.show_adj_close()
```

```python
# Plot the candle stick chart
AAPL.show_candle_stick()
```

```python
AAPL.show_line()
```
